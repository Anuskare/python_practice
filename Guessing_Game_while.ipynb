```python
from random import randint
a=randint(0,100)
guessno=[0]
print('''Welcome to the Guessing Game Challenge
The number would be between 1 and 100 including 1,100
In first turn, 
if your guess is within 10 of the number, return "WARM!"
further than 10 away from the number, return "COLD!"
On all subsequent turns, 
if a guess is closer to the number than the previous guess return "WARMER!"
if a guess is farther from the number than the previous guess, return "COLDER!"
When the player's guess equals the number, 
YOU HAVE WON!''')
while len(guessno)>0:
    g1=int(input("Please guess a number between 1 and 100 including 1,100 both "))
    guessno.append(a)
    if g1==a:
        print("YOU HAVE WON IN",len(guessno),"ATTEMPTS")
        break
    while len(guessno)==2:
        if (abs(g1-a)<10):
            print("Warm")
            break
        else:
            print("Cold")
            break
    while len(guessno)>2:
        if (abs(g1-a)<10):
            print("Warmer")
            break
        else:
            print("Colder")
            break
    
    
```
